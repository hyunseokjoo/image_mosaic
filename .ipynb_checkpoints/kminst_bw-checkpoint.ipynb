```python
import cv2, os
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

# Load and Resize Image

```python
img_path = 'img/cat.jpg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

img = cv2.resize(img, dsize=None, fx=0.2, fy=0.2) #fx fy  x y 80    .

print(img.shape)

plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(img, cmap='gray')
```

# Preview Patch Images

```python
''' 
npy  ?
numpy        npz  
numpy   .

npz  ?
numpy   npy    
npz  numpy   .
'''

sample_imgs = np.load('dataset/k49-train-imgs.npz')['arr_0'] 

plt.figure(figsize=(20,10))
for i in range(80):
    img_patch = 255 - sample_imgs[i]
    
    plt.subplot(5, 16, i+1)             #5 x 16  plot 
    plt.title(int(np.mean(img_patch)))  # 
    plt.axis('off')                     # 
    plt.imshow(img_patch, cmap='gray')  
```

# Distribution of Patch Images

```python
'''
 mnist dataset        255 

'''
means = np.mean(255 - sample_imgs, axis=(1,2)) 

plt.figure(figsize=(12, 6))
plt.hist(means, bins=50, log=True)
plt.show()
```

# Adjust MinMax of Input Image

```python
img = cv2.normalize(img, dst=None, alpha=120, beta=245, norm_type=cv2.NORM_MINMAX)

plt.figure(figsize=(12,6))
plt.hist(img.flatten(),bins=50, log=True)
plt.show()
```

# Organize Patch Images

```python
bins = defaultdict(list) # defaultdict dictionary      : https://dongdongfather.tistory.com/69

# bins       
#  120  120     
#   255  .
chk = 1 
for img_patch, mean in zip(sample_imgs, means):
    if chk == 1:
        print(img_patch)
        print(mean)
        chk += 1
    bins[int(mean)].append(img_patch)
    
print(len(bins))
```

# Fill Images

```python
'''
    
28*28     
'''
h, w = img.shape

img_out = np.zeros((h*28, w*28), dtype=np.uint8) #  28*28      

for y in range(h):
    for x in range(w):
        level = img[y,x]
        
        b = bins[level]
        
        while len(b) == 0:
            level += 1 
            b = bins[level]
            
        img_patch = 255 - b[np.random.randint(len(b))] #b     .
        
        img_out[y*28:(y+1)*28, x*28:(x+1)*28] = img_patch # output image   
        
plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(img_out, cmap='gray')

_ = cv2.imwrite('result/%s_bw.jpg' %os.path.splitext(os.path.basename(img_path))[0], img_out)

print(os.path.splitext(os.path.basename(img_path))[0])
```
