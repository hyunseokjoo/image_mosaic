```python
import cv2, os
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

# Load and Resize Image

```python
img_path = 'img/cat.jpg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

img = cv2.resize(img, dsize=None, fx=0.2, fy=0.2) #fx fy 설정하면 x축과 y축으로 80퍼 센트 줄인 이미지로 호출한다.

print(img.shape)

plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(img, cmap='gray')
```

# Preview Patch Images

```python
''' 
npy 파일 확장자란?
numpy 배열정보를 파일로 저장하기 위해 만든 파일 형식 npz 파일 안에는
numpy 배열정보가 들어 있다.

npz 파일 확장자란?
numpy 배열 정보가 npy는 하나만 가지고 있지만 
npz는 여러가의 numpy배열 정보가 담겨져 있다.
'''

sample_imgs = np.load('dataset/k49-train-imgs.npz')['arr_0'] 

plt.figure(figsize=(20,10))
for i in range(80):
    img_patch = 255 - sample_imgs[i]
    
    plt.subplot(5, 16, i+1)             #5 x 16 배열로 plot 설정
    plt.title(int(np.mean(img_patch)))  #배열의 중간값
    plt.axis('off')                     #축 없애기
    plt.imshow(img_patch, cmap='gray')  
```

# Distribution of Patch Images

```python
'''
원래 mnist dataset이 검정 배경이기 때문에 반대로 색반전을 주기 위해서 255에서 빼준다

'''
means = np.mean(255 - sample_imgs, axis=(1,2)) 

plt.figure(figsize=(12, 6))
plt.hist(means, bins=50, log=True)
plt.show()
```

# Adjust MinMax of Input Image

```python
img = cv2.normalize(img, dst=None, alpha=120, beta=245, norm_type=cv2.NORM_MINMAX)

plt.figure(figsize=(12,6))
plt.hist(img.flatten(),bins=50, log=True)
plt.show()
```

# Organize Patch Images

```python
bins = defaultdict(list) # defaultdict dictionary를 구현 하기 위한 함수 참조 : https://dongdongfather.tistory.com/69

# bins에는 평균값에 맞는 글자를 리스트 형식으로 넣어준다 
# 평균값이 120 이면 120에 해당하는 글자들을 다 넣는다 
# 이 작업을 255까지 진행 한다.
chk = 1 
for img_patch, mean in zip(sample_imgs, means):
    if chk == 1:
        print(img_patch)
        print(mean)
        chk += 1
    bins[int(mean)].append(img_patch)
    
print(len(bins))
```

# Fill Images

```python
'''
기존 이미지 픽셀 하나당 
28*28 이미지를 하나씩 대체해서 넣는 방식
'''
h, w = img.shape

img_out = np.zeros((h*28, w*28), dtype=np.uint8) #기존 픽셀에 28*28을 이미지를 넣기 위해서 높이와 너비에 곱해준다

for y in range(h):
    for x in range(w):
        level = img[y,x]
        
        b = bins[level]
        
        while len(b) == 0:
            level += 1 
            b = bins[level]
            
        img_patch = 255 - b[np.random.randint(len(b))] #b에 담긴 비스무리한 객체들중에 아무거나 삽입한다.
        
        img_out[y*28:(y+1)*28, x*28:(x+1)*28] = img_patch # output image에 픽셀마다 객체를 넣어준다
        
plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(img_out, cmap='gray')

_ = cv2.imwrite('result/%s_bw.jpg' %os.path.splitext(os.path.basename(img_path))[0], img_out)

print(os.path.splitext(os.path.basename(img_path))[0])
```
